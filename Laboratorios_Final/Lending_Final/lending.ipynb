```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuraci√≥n general
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")

# Cargar datos desde el archivo comprimido
df = pd.read_csv("accepted_2007_to_2018Q4.csv", low_memory=False)

# Mostrar las primeras filas del DataFrame
display(df.head())

# Ver columnas disponibles
print("\nColumnas en el dataset:")
print(df.columns.tolist())

# Tipos de datos y nulos
print("\nResumen de tipos de datos:")
print(df.dtypes)

print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# --- Punto e: Proporci√≥n de pagos completos vs impagos ---
# La columna 'loan_status' contiene esta informaci√≥n
print("\nDistribuci√≥n del estado del pr√©stamo (loan_status):")
print(df['loan_status'].value_counts(normalize=True) * 100)

# Visualizaci√≥n
df['loan_status'].value_counts().plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Distribuci√≥n de Estados del Pr√©stamo")
plt.xlabel("Estado")
plt.ylabel("Cantidad de Pr√©stamos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Punto f: Hip√≥tesis que explique la calificaci√≥n (grade) ---
# Hip√≥tesis: A mayor ingreso anual (annual_inc), mejor ser√° la calificaci√≥n del cr√©dito (grade).

# Filtrar columnas relevantes y eliminar nulos
grade_income_df = df[['grade', 'annual_inc']].dropna()

# Boxplot para observar la distribuci√≥n de ingresos por calificaci√≥n
plt.figure(figsize=(10,6))
sns.boxplot(x='grade', y='annual_inc', data=grade_income_df, order=sorted(df['grade'].dropna().unique()))
plt.yscale('log')  # Escala logar√≠tmica por valores extremos en ingresos
plt.title("Distribuci√≥n del Ingreso Anual por Calificaci√≥n del Cr√©dito")
plt.xlabel("Calificaci√≥n (Grade)")
plt.ylabel("Ingreso Anual (escala log)")
plt.tight_layout()
plt.show()

# Mostrar ingreso promedio por calificaci√≥n
print("\nIngreso promedio anual por calificaci√≥n:")
print(grade_income_df.groupby('grade')['annual_inc'].mean().sort_index())
```

```python
# --- Punto g: Relaci√≥n entre calificaci√≥n (grade), pagos (loan_status) y prop√≥sito (purpose) ---
# Filtrar columnas necesarias y eliminar nulos
c_df = df[['grade', 'loan_status', 'purpose']].dropna()

# Agrupar y contar combinaciones
relacion = c_df.groupby(['grade', 'loan_status']).size().unstack(fill_value=0)

# Visualizaci√≥n: Estado del pr√©stamo por calificaci√≥n
relacion.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Paired')
plt.title("Relaci√≥n entre Calificaci√≥n y Estado del Pr√©stamo")
plt.xlabel("Calificaci√≥n (Grade)")
plt.ylabel("Cantidad de Pr√©stamos")
plt.tight_layout()
plt.show()

# Relaci√≥n entre prop√≥sito y calificaci√≥n
plt.figure(figsize=(12,6))
sns.countplot(data=c_df, x='purpose', hue='grade', order=c_df['purpose'].value_counts().index)
plt.title("Distribuci√≥n del Prop√≥sito del Pr√©stamo por Calificaci√≥n")
plt.xlabel("Prop√≥sito del Pr√©stamo")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Punto h: Relaci√≥n entre calificaci√≥n (grade) y sub-calificaci√≥n (sub_grade) ---
# Contar frecuencia de sub_grades por cada grade
plt.figure(figsize=(12,6))
sns.countplot(data=df.dropna(subset=['grade', 'sub_grade']), x='sub_grade', hue='grade',
              order=sorted(df['sub_grade'].dropna().unique()))
plt.title("Distribuci√≥n de Sub-Calificaciones por Calificaci√≥n")
plt.xlabel("Sub-Calificaci√≥n (sub_grade)")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Punto i: Relaci√≥n entre calificaci√≥n y raz√≥n de ingreso deuda (dti) ---
i_df = df[['grade', 'dti']].dropna()

plt.figure(figsize=(10,6))
sns.boxplot(data=i_df, x='grade', y='dti', order=sorted(i_df['grade'].unique()))
plt.title("Distribuci√≥n de la Raz√≥n Deuda-Ingreso (DTI) por Calificaci√≥n")
plt.xlabel("Calificaci√≥n (Grade)")
plt.ylabel("DTI")
plt.tight_layout()
plt.show()

# Media de DTI por calificaci√≥n
print("\nMedia del DTI por calificaci√≥n:")
print(i_df.groupby('grade')['dti'].mean().sort_index())

# --- Cr√©ditos por categor√≠a y an√°lisis del monto del pr√©stamo ---
# Filtrar columnas relevantes y eliminar nulos
purpose_df = df[['purpose', 'loan_amnt']].dropna()

# Conteo de pr√©stamos por categor√≠a
print("\nCantidad de cr√©ditos por categor√≠a (purpose):")
print(purpose_df['purpose'].value_counts())

# Visualizaci√≥n del conteo
plt.figure(figsize=(12,6))
sns.countplot(data=purpose_df, x='purpose', order=purpose_df['purpose'].value_counts().index, palette='Set2')
plt.title("Cantidad de Cr√©ditos por Categor√≠a")
plt.xlabel("Prop√≥sito del Pr√©stamo")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Boxplot del monto del pr√©stamo por categor√≠a
plt.figure(figsize=(14,6))
sns.boxplot(data=purpose_df, x='purpose', y='loan_amnt', order=purpose_df['purpose'].value_counts().index, palette='Set3')
plt.title("Distribuci√≥n del Monto del Pr√©stamo por Categor√≠a")
plt.xlabel("Prop√≥sito del Pr√©stamo")
plt.ylabel("Monto del Pr√©stamo (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Estad√≠sticas resumen por categor√≠a
print("\nEstad√≠sticas del monto del pr√©stamo por categor√≠a:")
print(purpose_df.groupby('purpose')['loan_amnt'].describe())
```

```python
#J
#Preparar la informaci√≥n
rate_df = df[['int_rate', 'purpose', 'sub_grade']].dropna()
rate_df['int_rate'] = rate_df['int_rate'].astype(str).str.rstrip('%').astype(float)

#Agrupar y calcular estadisticas
rate_stats = rate_df.groupby(['purpose', 'sub_grade'])['int_rate'].agg(
    promedio='mean',
    desviacion='std',
    minimo='min',
    maximo='max'
).reset_index()

# Mostramos las estad√≠sticas
print("\nüìä Estad√≠sticas de tasa de inter√©s por categor√≠a y subcategor√≠a:")
display(rate_stats.head(20))

#Visualizar los resultados
plt.figure(figsize=(14,7))
sns.boxplot(data=rate_df, x='sub_grade', y='int_rate', hue='purpose')
plt.title("Distribuci√≥n de Tasa de Inter√©s por Sub-Calificaci√≥n y Categor√≠a")
plt.xlabel("Sub-Calificaci√≥n (sub_grade)")
plt.ylabel("Tasa de Inter√©s (%)")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#K
#Prearar los datos
# Selecci√≥n y limpieza
purpose_analysis = df[['purpose', 'loan_amnt', 'term', 'loan_status']].dropna()

# Limpieza de texto en 'term' (elimina " months")
purpose_analysis['term'] = purpose_analysis['term'].str.extract('(\d+)').astype(int)

#Resumen estadistico por proposito
resumen = purpose_analysis.groupby('purpose').agg(
    cantidad_creditos=('loan_amnt', 'count'),
    monto_total=('loan_amnt', 'sum'),
    monto_promedio=('loan_amnt', 'mean'),
    plazo_promedio=('term', 'mean')
).sort_values(by='cantidad_creditos', ascending=False)

print("üìä Resumen de montos, plazos y cantidad de cr√©ditos por prop√≥sito:")
display(resumen)

#Tasa de incumplimiento por proposito
# Crear columna binaria de incumplimiento
purpose_analysis['incumplimiento'] = ~purpose_analysis['loan_status'].isin(['Fully Paid', 'Current'])

# Tasa de incumplimiento por prop√≥sito
incumplimiento_por_purpose = purpose_analysis.groupby('purpose')['incumplimiento'].mean().sort_values(ascending=False)

print("\nüìâ Tasa de incumplimiento por prop√≥sito (%):")
print((incumplimiento_por_purpose * 100).round(2))

#Visualizaciones
#Creditos por Proposito
plt.figure(figsize=(12,5))
sns.countplot(data=purpose_analysis, x='purpose', order=purpose_analysis['purpose'].value_counts().index, palette='Set2')
plt.title("Cantidad de Cr√©ditos por Prop√≥sito")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Tasa de Incumplimiento por proposito
plt.figure(figsize=(12,5))
incumplimiento_por_purpose.plot(kind='bar', color='salmon')
plt.title("Tasa de Incumplimiento por Prop√≥sito")
plt.ylabel("Tasa de Incumplimiento (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#I

# Filtrar cr√©ditos que fueron pagados completamente
pagados_df = df[df['loan_status'] == 'Fully Paid'].copy()

# Usamos las fechas para calcular duraci√≥n real del cr√©dito
pagados_df = pagados_df[['issue_d', 'last_pymnt_d', 'term', 'purpose']].dropna()

# Convertir columnas de fecha al formato datetime
pagados_df['issue_d'] = pd.to_datetime(pagados_df['issue_d'], format='%b-%Y')
pagados_df['last_pymnt_d'] = pd.to_datetime(pagados_df['last_pymnt_d'], format='%b-%Y')

# Calcular duraci√≥n en meses (diferencia entre fechas)
pagados_df['duracion_real_meses'] = ((pagados_df['last_pymnt_d'] - pagados_df['issue_d']) / pd.Timedelta(days=30)).round().astype(int)

# Extraer plazo oficial como n√∫mero entero
pagados_df['plazo_estipulado'] = pagados_df['term'].str.extract('(\d+)').astype(int)

#Resumen Estadistico
print("üìä Estad√≠sticas de duraci√≥n real del pr√©stamo (en meses):")
print(pagados_df['duracion_real_meses'].describe())

# ¬øCoincide con el plazo estipulado?
comparacion = pagados_df[['duracion_real_meses', 'plazo_estipulado']]
print("\nüìà Comparaci√≥n entre duraci√≥n real y plazo estipulado (algunas filas):")
print(comparacion.sample(10))

#Analisis por proposito
resumen_purpose = pagados_df.groupby('purpose')['duracion_real_meses'].describe().sort_values(by='mean', ascending=False)

print("\nüìä Estad√≠sticas por prop√≥sito (duraci√≥n real del cr√©dito en meses):")
display(resumen_purpose)

#Visualizaci√≥n en Boxplot
plt.figure(figsize=(14,6))
sns.boxplot(data=pagados_df, x='purpose', y='duracion_real_meses', order=pagados_df['purpose'].value_counts().index)
plt.title("Duraci√≥n Real de los Cr√©ditos por Prop√≥sito")
plt.xlabel("Prop√≥sito del Pr√©stamo")
plt.ylabel("Duraci√≥n Real (meses)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# ¬øCu√°ndo los datos faltantes SON NORMALES en el negocio?
# Caso: months_since_last_delinquency (mths_since_last_delinq)
# Caso: emp_length -> longitud de empleo (puede faltar si est√° desempleado, trabaja informalmente o no quiso declararlo)

# --- Tratamiento de datos faltantes con sentido de negocio ---

# Copia de trabajo
df_missing = df.copy()

# 1. mths_since_last_delinq
# Reemplazar NaN por -1 indicando que nunca ha tenido morosidad
df_missing['mths_since_last_delinq_filled'] = df_missing['mths_since_last_delinq'].fillna(-1)

# 2. emp_length
# Reemplazar nulos por "Desconocido"
df_missing['emp_length_filled'] = df_missing['emp_length'].fillna("Desconocido")

# 3. joint variables: solo se imputan si el pr√©stamo es conjunto
df_missing['annual_inc_joint_filled'] = df_missing.apply(
    lambda row: row['annual_inc_joint'] if row['application_type'] == 'Joint App' else row['annual_inc'],
    axis=1
)

# 4. revol_util: imputar con la mediana si es necesario
if df_missing['revol_util'].dtype == 'object':
    revol_util_numeric = pd.to_numeric(df_missing['revol_util'].str.rstrip('%'), errors='coerce')
else:
    revol_util_numeric = pd.to_numeric(df_missing['revol_util'], errors='coerce')

median_revol_util = revol_util_numeric.median()
df_missing['revol_util_filled'] = revol_util_numeric.fillna(median_revol_util)

# Verificar imputaciones
print("\nResumen de imputaciones:")
print("mths_since_last_delinq_filled - valores √∫nicos:", df_missing['mths_since_last_delinq_filled'].unique()[:5])
print("emp_length_filled - valores √∫nicos:", df_missing['emp_length_filled'].unique()[:5])
print("annual_inc_joint_filled - nulos:", df_missing['annual_inc_joint_filled'].isnull().sum())
print("revol_util_filled - estad√≠stica:", df_missing['revol_util_filled'].describe())
```

```python
#P
#¬øCu√°les son los datos que deben estar disponibles cuando se desea invertir en un cr√©dito?
# Lista de variables clave para invertir en un cr√©dito
columnas_clave = [
    'loan_amnt',             # Monto del pr√©stamo
    'term',                  # Plazo
    'int_rate',              # Tasa de inter√©s
    'grade', 'sub_grade',    # Categor√≠a y subcategor√≠a
    'purpose',               # Prop√≥sito del cr√©dito
    'emp_length',            # Longitud del empleo
    'annual_inc',            # Ingreso anual
    'home_ownership',        # Tipo de vivienda
    'dti',                   # Deuda sobre ingresos
    'fico_range_low', 'fico_range_high',  # Rango de score crediticio
    'revol_util',            # Utilizaci√≥n del cr√©dito rotativo
    'mths_since_last_delinq',# Tiempo desde √∫ltima morosidad
    'application_type',      # Tipo de solicitud (Individual/Conjunta)
    'annual_inc_joint',      # Ingreso conjunto (si aplica)
    'loan_status',           # Estado del pr√©stamo (si aplica)
    'issue_d',               # Fecha de emisi√≥n
    'total_pymnt',           # Pagos realizados (si aplica)
    'last_pymnt_d'           # Fecha del √∫ltimo pago (si aplica)
]

# Crear subconjunto con los datos relevantes
df_inversion = df[columnas_clave].copy()

# Visualizar una muestra del nuevo DataFrame con los campos clave
print("Estructura de entrada para evaluaci√≥n de cr√©dito:")
print(df_inversion.head())
```
